- ***data understanding***

- **name: The name or model of the car being sold.**
- **year: The manufacturing year of the car.**
- **selling_price: The price at which the car is being sold.**
- **km_driven: The distance the car has been driven in kilometers.**
- **fuel: The type of fuel used by the car (e.g., Diesel, Petrol).**
- **seller_type: The type of seller selling the car (e.g., Individual, Dealer).**
- **transmission: The type of transmission of the car (e.g., Manual, Automatic).**
- **owner: The ownership history of the car (e.g., First Owner, Second Owner).**
- **mileage: The fuel efficiency or mileage of the car.**
- **engine: The engine displacement of the car in cubic centimeters (CC).**
- **max_power: The maximum power output of the engine in brake horsepower (bhp).**
- **seats: The number of seats in the car.**
